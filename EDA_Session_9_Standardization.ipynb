**Normalization**

**task**
  
- read the data again
- step-1: take prevailing wage column: **visa_df['prevailing_wage']**
- step-2: calculate the min value prevailing wage: **visa_df['prevailing_wage'].min**
- step-3: calculate the max value prevailing wage: **visa_df['prevailing_wage'].max**

- step-4: calculate the Nr: Step-1 - Step-2
- step-5:  Dr = step3/step2
- step-6: divide the step4/step5



```python
min = visa_df['prevailing_wage'].min()

max = visa_df['prevailing_wage'].max()

Nr = visa_df['prevailing_wage']-min

visa_df['prevailing_wage_min_max'] = Nr/(max-min)
visa_df[['prevailing_wage','prevailing_wage_min_max']]

```
