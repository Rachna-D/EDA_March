**Data Transformations**

- Generally, data has 3 types of distribution
    - Normal distribution
    - Positive Skewed
    - Negative Skewed

- All maths concepts developed by assumption as Data follows Normal distribution
- Data skew happens because of many reasons, in that one is Outliers
- Once, we perform the outlier analysis then we can apply transformation
methods, if data does not follow normality.
- we have some transformation methods mentioned below
    - Log Transformation
    - Exponential Transformation
    - Reciprocal Transformation
    - Square root Transformation
    - Power Transformation

$Step-1$

**import the packages**

```python
import numpy as np
import matplotlib.pyplot as plt
```

$Step-2$

**Read the data**

```python
exp_data=np.random.exponential(size=10000)
exp_data
```

$Step-3$

**Plot the data**

```python
plt.hist(exp_data,bins=20,label='Exponential')
plt.show()
```

```python
norm_data=np.random.normal(size=10000)
plt.hist(norm_data,bins=20,label='Exponential Data')
plt.legend()
plt.show()
```

**Goal**

**Convert exp_data to Normal distribution**

bins means intervals

$Step-4$

**Log transformation**

  - Log transformation means performing logarithm operation on original data
- it is as approach to convert into Normal distrubution
- Log means natural logarithm - base=e

```python
x=2
np.log(x)
```

```python
log_data=np.log(exp_data)
log_data
```

```python
plt.hist(log_data,bins=20,label='Logarithm Data')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1).hist(exp_data,bins=30,label='Exponential')
plt.legend()
plt.subplot(1,2,2).hist(log_data,bins=30,label='Logarithm')
plt.legend()
plt.show()
```

**Reciprocal Transformations**

- Reciprocal will fail when values has low
- For e.g. x=0, then 1/0 fail

```python
x=2
1/x
```

```python
reciprocal_data=np.reciprocal(exp_data)
plt.hist(reciprocal_data,bins=20,label='Reciprocal Data')
plt.legend()
plt.show()
```

**Square root Transformation**

```python
squareroot=np.sqrt(exp_data)
plt.hist(squareroot,bins=20,label='Square root Data')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,4))

plt.subplot(2,2,1).hist(exp_data,bins=20,label='Exponential')
plt.legend()
plt.subplot(2,2,2).hist(log_data,bins=20,label='Logarithm')
plt.legend()
plt.subplot(2,2,3).hist(reciprocal_data,bins=20,label='Reciprocal')
plt.legend()
plt.subplot(2,2,4).hist(squareroot,bins=20,label='Square root')
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.subplot(1,3,1).hist(exp_data,bins=20,label='Exponential')
plt.legend()
plt.subplot(1,3,2).hist(log_data,bins=20,label='Logarithm')
plt.legend()
plt.subplot(1,3,3).hist(squareroot,bins=20,label='Square root')
plt.legend()

plt.show()
```

**Power Transformation**

- Box cox
- ye-johnson
- it is under sklearn preprocessing

```python

from sklearn.preprocessing import PowerTransformer
# read the package
# save the package
# apply fit tranform

power=PowerTransformer(method='yeo-johnson')
exp_data1=exp_data.reshape(-1,1)
data_trans=power.fit_transform(exp_data1)
plt.hist(data_trans)
```

```python
from sklearn.preprocessing import PowerTransformer

power=PowerTransformer(method='yeo-johnson')
data_trans=power.fit_transform(exp_data)
plt.hist(data_trans)
```
