**percentile and quantile**

- percentile
    - np.percentile()
    - it takes two arguments
        - data : a
        - percentile : q -- the values vary from 0 to 100
        - if you want 50p data then q=50

- quantile
    - np.quantile()
    - it takes two arguments
        - data : a
        - percentile : q -- the values vary from 0 to 100
        - if you want 50p data then q=0.5

**Empirical rule (68-95-99.7)**



- first calculate mean value
- second calculate std value
- con1: mean-1*std
- con2: mean-2*std
- if you apply above conditions on wage data, output count should be equal to 68 percentile data
- 68 % total employee: 17326

```python
v1=wage_mean-1*wage_std
v2=wage_mean+1*wage_std
v1,v2

con1=visa_df['prevailing_wage']>v1
con2=visa_df['prevailing_wage']<v2

count1=visa_df[con1 & con2]
count1
```

```python
## All together

l1=[]
cols=['no_of_employees','yr_of_estab','prevailing_wage']

##also cols=visa_df.select_dtypes(exclude='object').columns

for i in cols:
    countt=round(len(visa_df[i]),2)
    mean=round(visa_df[i].mean(),2)
    median=round(visa_df[i].median(),2)
    maxx=round(visa_df[i].max(),2)
    minx=round(visa_df[i].min(),2)
    std=round(visa_df[i].std(),2)
    
    ###################################################################
    
    p_25=round(np.quantile(visa_df[i],0.25),2)
    p_50=round(np.quantile(visa_df[i],0.50),2)
    p_75=round(np.quantile(visa_df[i],0.75),2)
    
    ###################################################################
    
    l1.append([countt,mean,median,maxx,minx,std,p_25,p_50,p_75])

l1

index_val=['count','mean','median','maxx','minx','std','p_25','p_50','p_75']
df=pd.DataFrame(l1,
               columns=index_val,
               index=cols).T

df
```
