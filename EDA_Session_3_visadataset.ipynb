$take$

- take is used to extract the particular index of data
- take one parameter called axis
- axis=1 represents columns
- axis=0 represents rows
- by default axis=0 is there
- in below e.g. we ggiven 100,200,300 & we didn't provide axis values
- it understands we are talking about rows

```python
visa_df.take([100,200,300])
```

```python
visa_df.take([100,200,300],axis=1)
```

```python
visa_df.take([11],axis=1)
```

```python
visa_df.take([100:105])  ## Not working
```

```python
## I want rows & columns

visa_df.take([100]).take([2],axis=1)
```
